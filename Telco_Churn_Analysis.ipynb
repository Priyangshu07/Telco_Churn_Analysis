```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('Customer Churn.csv')
df
```

```python
df.info()
```

#replaced the blanks with 0 as tenure is 0 and not total charges are recorded for them 

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)

df.head()
```

```python
ax=sns.countplot(x='Churn',data=df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb=df.groupby("Churn").agg({'Churn':'count'})

plt.pie(gb['Churn'], labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned Customers",fontsize=10)

plt.show()
```

From the above Pie Chart we can discover that 26.54% of our customers have churned out.
#Now we can explore the reason behind it 

```python
plt.figure(figsize=(5,5))
sns.countplot(x='gender',data=df,hue='Churn')
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,bins=72,hue='Churn')
plt.show()
```

**People who have used our services for a long time have stayed and people who have used our services for 1 or 2 months have churned

```python
plt.figure(figsize=(5,6))
ax=sns.countplot(x='Contract', data=df, hue='Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

**People who have Month-to-Month contracts are more likely to churn then from those who have 1 or 2 years of contracts.

```python
df.columns.values
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(['PhoneService', 'MultipleLines', 'InternetService',
                         'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                         'TechSupport', 'StreamingTV', 'StreamingMovies']):
    sns.countplot(data=df, x=col, ax=axes[i],hue='Churn')
    axes[i].set_title(f'Countplot of {col}')

plt.tight_layout()
plt.show()
```

For most services like PhoneService, MultipleLines, OnlineSecurity, and TechSupport, customers who did not churn (blue) are significantly more frequent than those who did (orange). In categories like InternetService, churn rates are higher for customers using Fiber optic than DSL. There’s a visible trend showing that customers without optional services (e.g., OnlineBackup, DeviceProtection) tend to churn more.
